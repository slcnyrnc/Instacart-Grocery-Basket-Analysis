```python
# Defining price range function

def price_label(row):
    if row['prices'] <= 5:
        return 'Low-range product'
    elif (row['prices'] > 5) and (row['prices'] <= 15):
        return 'Mid-range product'
    elif row['prices'] > 15:
        return 'High-range product'
    else: return 'Not enough data'
```

```python
# Apply function

df['price_range'] = df.apply(price_label, axis=1)
```

#### The following part, I found online because I wasn't sure about how to interpret "busiest hours".

```python
percentiles = np.percentile(df_ords_prods_merged['order_hour_of_day'], [25, 75])
```

```python
most_orders_hours = df_ords_prods_merged['order_hour_of_day'] >= percentiles[1]
fewest_orders_hours = df_ords_prods_merged['order_hour_of_day'] <= percentiles[0]
average_orders_hours = ~most_orders_hours & ~fewest_orders_hours
```

```python
df_ords_prods_merged['hour_order_volume'] = np.where(most_orders_hours, 'Most orders',
                                                     np.where(fewest_orders_hours, 'Fewest orders', 'Average orders'))
```
